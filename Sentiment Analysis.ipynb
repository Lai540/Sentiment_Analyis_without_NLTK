```python
#first we must clean the text to be analyzed.

sentiment = open('text.txt', encoding = 'utf-8').read()

print (sentiment)
```

```python
lower = sentiment.lower()
```

```python
lower
```

```python
import string
```

```python
#Let us print the available special charaters that we can remove from the text.
print("These are the special Characters:\n\n", string.punctuation)
```

```python
#Next let us now remove the punctuations

cleaned_text = lower.translate(str.maketrans('','', string.punctuation))
```

```python
cleaned_text
```

```python
#Now the next step is to tokenized the text: This is the conversion of text to words.
#words are stored in a list
tokenized_text = cleaned_text.split()
```

```python
tokenized_text
```

```python
#Let's remove some of the possible stop words, meaningless words in the context of NLP

stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now", "id", "us", "art"]
fully_cleaned = []
for word in tokenized_text:
    if word not in stop_words:
        fully_cleaned.append(word)
        
fully_cleaned
```

```python
emotion_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", "").strip()
        # Checking if the line contains a colon
        if ':' in clear_line:
            word, emotion = clear_line.split(':')
            if word in fully_cleaned:
                emotion_list.append(emotion)
    else:
        print(f"Issue with line: {clear_line}")  # Print lines that don't match the expected format for debugging
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Your Counter with emotions
emotions_counter = Counter({
    'fearful': 1,
    'safe': 3,
    'happy': 1,
    'anxious': 1
})

# Get the total count of emotions
total_emotions = sum(emotions_counter.values())

# Calculate percentages for each emotion
emotions_percentages = {emotion: (count / total_emotions) * 100 for emotion, count in emotions_counter.items()}

# Create lists for plotting
emotions = list(emotions_percentages.keys())
percentages = list(emotions_percentages.values())

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=emotions, autopct='%1.1f%%', startangle=140)
plt.title('Emotions Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Extract emotions and counts
emotions = list(emotions_counter.keys())
counts = list(emotions_counter.values())

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(emotions, counts, color='skyblue')
plt.xlabel('Emotions')
plt.ylabel('Counts')
plt.title('Distribution of Emotions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```
